date: La fecha en la que se registró la venta de la casa.

price: El precio de venta de la casa.

bedrooms: El número de dormitorios en la casa.

bathrooms: El número de baños en la casa.

sqft_living: El tamaño (área) en pies cuadrados del espacio habitable (área interna) de la casa.

sqft_lot: El tamaño del terreno en pies cuadrados en el que se encuentra la casa.

floors: El número de pisos o niveles en la casa.

waterfront: Un valor binario que indica si la propiedad tiene frente de agua (1 si es verdadero, 0 si es falso).

view: Un índice que indica la calidad de la vista desde la propiedad.

grade: Un índice que representa la calidad general de la construcción y el diseño de la casa.

sqft_above: El tamaño en pies cuadrados del espacio habitable (área interna) en la parte superior de la planta baja.

sqft_basement: El tamaño en pies cuadrados del área del sótano de la casa.

yr_built: El año en que se construyó la casa.

yr_renovated: El año en que se realizó una renovación importante en la casa, si es el caso.

zipcode: El código postal de la ubicación de la casa.

lat: La latitud de la ubicación de la casa.

long: La longitud de la ubicación de la casa.

sqft_living15: El tamaño promedio en pies cuadrados del espacio habitable de las 15 casas más cercanas.

sqft_lot15: El tamaño promedio en pies cuadrados del terreno de los 15 lotes más cercanos.

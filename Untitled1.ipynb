<h1>Información de la Casa</h1>
    <ul>
        <li><strong>date:</strong> La fecha en la que se registró la venta de la casa.</li>
        <li><strong>price:</strong> El precio de venta de la casa.</li>
        <li><strong>bedrooms:</strong> El número de dormitorios en la casa.</li>
        <li><strong>bathrooms:</strong> El número de baños en la casa.</li>
        <li><strong>sqft_living:</strong> El tamaño (área) en pies cuadrados del espacio habitable (área interna) de la casa.</li>
        <li><strong>sqft_lot:</strong> El tamaño del terreno en pies cuadrados en el que se encuentra la casa.</li>
        <li><strong>floors:</strong> El número de pisos o niveles en la casa.</li>
        <li><strong>waterfront:</strong> Un valor binario que indica si la propiedad tiene frente de agua (1 si es verdadero, 0 si es falso).</li>
        <li><strong>view:</strong> Un índice que indica la calidad de la vista desde la propiedad.</li>
        <li><strong>grade:</strong> Un índice que representa la calidad general de la construcción y el diseño de la casa.</li>
        <li><strong>sqft_above:</strong> El tamaño en pies cuadrados del espacio habitable (área interna) en la parte superior de la planta baja.</li>
        <li><strong>sqft_basement:</strong> El tamaño en pies cuadrados del área del sótano de la casa.</li>
        <li><strong>yr_built:</strong> El año en que se construyó la casa.</li>
        <li><strong>yr_renovated:</strong> El año en que se realizó una renovación importante en la casa, si es el caso.</li>
        <li><strong>zipcode:</strong> El código postal de la ubicación de la casa.</li>
        <li><strong>lat:</strong> La latitud de la ubicación de la casa.</li>
        <li><strong>long:</strong> La longitud de la ubicación de la casa.</li>
        <li><strong>sqft_living15:</strong> El tamaño promedio en pies cuadrados del espacio habitable de las 15 casas más cercanas.</li>
        <li><strong>sqft_lot15:</strong> El tamaño promedio en pies cuadrados del terreno de los 15 lotes más cercanos.</li>
    </ul>

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import cross_val_score, KFold
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import max_error
from sklearn.metrics import mean_squared_error, mean_absolute_error, max_error, r2_score, make_scorer
```

<h1>Validacion cruzada</h1>

```python
x = df.drop(columns=['price'], axis=1)
y = df['price']

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(x)
```

<h1>RandomForestRegressor</h1>

```python
rf_model = RandomForestRegressor(random_state=42)
kfold = KFold(n_splits=5, shuffle=True, random_state=42)
mse_scores = -cross_val_score(rf_model, X_scaled, y, cv=kfold, scoring='neg_mean_squared_error')
mae_scores = -cross_val_score(rf_model, X_scaled, y, cv=kfold, scoring='neg_mean_absolute_error')
r2_scores = cross_val_score(rf_model, X_scaled, y, cv=kfold, scoring='r2')

mean_mse_rf = mse_scores.mean()
mean_mae_rf = mae_scores.mean()
mean_r2_rf = r2_scores.mean()

print(f"Media del Error Cuadrático Medio (MSE): {mean_mse_rf}")
print(f"Media del Error Absoluto Medio (MAE): {mean_mae_rf}")
print(f"Media de R-squared (R2): {mean_r2_rf}")
```

<h1>AdaBoostRegressor</h1>

```python
ab_model = AdaBoostRegressor(random_state=42)
kfold = KFold(n_splits=5, shuffle=True, random_state=42)
mse_scores = -cross_val_score(ab_model, X_scaled, y, cv=kfold, scoring='neg_mean_squared_error')
mae_scores = -cross_val_score(ab_model, X_scaled, y, cv=kfold, scoring='neg_mean_absolute_error')
r2_scores = cross_val_score(ab_model, X_scaled, y, cv=kfold, scoring='r2')

mean_mse_ab = mse_scores.mean()
mean_mae_ab = mae_scores.mean()
mean_r2_ab = r2_scores.mean()

print(f"Media del Error Cuadrático Medio (MSE): {mean_mse_ab}")
print(f"Media del Error Absoluto Medio (MAE): {mean_mae_ab}")
print(f"Media de R-squared (R2): {mean_r2_ab}")
```

<h1>XGBRegressor</h1>

```python
xgb_model = XGBRegressor(random_state=42)
kfold = KFold(n_splits=5, shuffle=True, random_state=42)
mse_scores = -cross_val_score(xgb_model, X_scaled, y, cv=kfold, scoring='neg_mean_squared_error')
mae_scores = -cross_val_score(xgb_model, X_scaled, y, cv=kfold, scoring='neg_mean_absolute_error')
r2_scores = cross_val_score(xgb_model, X_scaled, y, cv=kfold, scoring='r2')

mean_mse_xgb = mse_scores.mean()
mean_mae_xgb = mae_scores.mean()
mean_r2_xgb = r2_scores.mean()

print(f"Media del Error Cuadrático Medio (MSE): {mean_mse_xgb}")
print(f"Media del Error Absoluto Medio (MAE): {mean_mae_xgb}")
print(f"Media de R-squared (R2): {mean_r2_xgb}")
```
